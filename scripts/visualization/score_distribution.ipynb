```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../../data/system_prompts/Prompt-Scores_Good-Property.csv")
data["Property_Num"] = data["Comments"].apply(lambda x:int(x.split()[1]))
data
```

```python
sns.lmplot(data=data, x="Property_Num", y="MMLU_acc")
```

```python
sns.kdeplot(data=data, x="MMLU_acc", fill=True, alpha=0)
plt.show()
```

```python
for var in data["Property_Num"].unique():
    sns.kdeplot(data=data[data["Property_Num"] == var], x="MMLU_acc", fill=True, alpha=0, label=var)
plt.legend()
plt.show()
```
