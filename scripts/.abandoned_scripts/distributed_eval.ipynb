```python
score_table = pd.read_csv("./data/system_prompts/Prompt-Scores_Good-Property.csv")
```

```python
raw = pd.read_csv("./results/benchmark/Mixtral-8x7B-Instruct-v0.1_truthfulqa_results.csv")
```

```python
def scoring_fn(idx, alist):
    return (idx, len(alist))
mp_list = [(idx, raw[prompt]) for idx, prompt in enumerate(score_table["Prompt"])]
pool = Pool(os.cpu_count())
x = pool.starmap(scoring_fn, mp_list)
for y in x:
    print(y)
```

```python
from datasets import load_metric
```

```python
bleurt = load_metric("bleurt")
```

```python
benchmark_obj.eval_saved_file("./results/benchmark/Mixtral-8x7B-Instruct-v0.1_truthfulqa_results.csv", "./data/system_prompts/Prompt-Scores_Good-Property.csv", metric_list=["bleu"])
```

```python
for prompt in score_table["Prompt"]:
    print(lib.utils.bleu_score(raw[prompt], benchmark_obj.correct_answer_list, benchmark_obj.incorrect_answer_list))
```

```python
for prompt in tqdm(score_table["Prompt"]):
    print(lib.utils.bleurt_score(raw[prompt], benchmark_obj.correct_answer_list, benchmark_obj.incorrect_answer_list, bleurt))
```
